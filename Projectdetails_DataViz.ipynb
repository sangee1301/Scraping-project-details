```python
#importing necessary libraries
import pandas as pd
import numpy as np

import seaborn as sns
sns.set_style('whitegrid')
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
#importing from csv file to a dataframe
details_df = pd.read_csv("windfarms_project_details.csv", index_col = False)
```

```python
details_df.head(10)
```

```python
ProjectName = details_df[details_df['1']=='Name']['2']
Country = details_df[details_df['0']=='Country name']['1']
Location = details_df[details_df['0']=='Location & Environment']['2']
Region = details_df[details_df['0']=='Region']['1']
Status = details_df[details_df['0']=='Development status']['1']
ProjectCapacity_MW = details_df[details_df['1']=='Project capacity']['2'].str.extract('(\d+)', expand=True).astype(float)
TurbineModel = details_df[details_df['0']=='Turbine model']['1']
TurbineCapacity_MW = details_df[details_df['0']=='Turbine capacity']['1'].str.extract('(\d+)', expand=True).astype(float)
No_of_Turbines = details_df[details_df['0']=='Number of turbines']['1']
Cost = details_df[details_df['0']=='Costs']['2']
Area_km2 = details_df[details_df['0']=='Area']['1'].str.extract('(\d+)', expand=True).astype(float)
Distance_From_Shore = details_df[details_df['0']=='Distance from shore (computed from centre)']['1'].str.extract('(\d+)', expand=True).astype(float)
Co2_reduced_per_year_tonnes = details_df[details_df['0']=='CO2 reduced per year']['1'].str.extract('(\d+)', expand=True).astype(float)
So2_reduced_per_year_tonnes = details_df[details_df['0']=='SO2 reduced per year']['1'].str.extract('(\d+)', expand=True).astype(float)
```

## Countplot by location of projects

```python
plt.figure(figsize=(10,4))
Location.value_counts().head(15).plot(kind='bar')
```

It looks like maximum number of projects are from the North Sea Location and the least from Pacific Ocean

## Count of projects by their status

```python
plt.figure(figsize=(8,4))
Status.value_counts().plot(kind='bar')
```

```python
ProjectCapacity_MW.plot(kind ='kde', figsize=(9,5))
```

## Histogram of Turbine capacity

```python
TurbineCapacity_MW.plot.hist(bins=10, figsize=(9,5))
```

```python
Area_km2.plot.area(figsize=(13,5), cmap= 'plasma', alpha =0.6)
```

```python
Co2_reduced_per_year_tonnes.plot(kind='kde', figsize=(10,5))
```
